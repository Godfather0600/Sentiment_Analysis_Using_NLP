```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to your metrics file
file_path = "../models/all_model/metrics.json"

with open(file_path, "r") as f:
    metrics = json.load(f)

metrics
```

```python
# Convert dictionary to DataFrame
df = pd.DataFrame(metrics).T

# Display comparison table
df
```

```python
# Save comparison table
df.to_csv("../models/model_comparison.csv")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df.index, df["accuracy"])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
metrics_list = ["accuracy", "precision", "recall", "f1-score"]

for metric in metrics_list:
    plt.figure(figsize=(8, 5))
    plt.bar(df.index, df[metric])
    plt.title(f"Model Comparison - {metric.capitalize()}")
    plt.ylabel(metric.capitalize())
    plt.xlabel("Model")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

```python
best_models = df.idxmax()
best_models
```

```python
print("Best models by metric:")
for metric in df.columns:
    print(f"{metric}: {df[metric].idxmax()}")
```
